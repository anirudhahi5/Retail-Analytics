### Here we see that most of the customer are from AU. So we can filter out AU customer

```python
AU_data=data[data.Country=='AU']
```

```python
AU_data.info()
```

```python
AU_data.head()
```

```python
AU_data.describe(include='all')
```

```python
AU_data=AU_data[['user ID','timestamp','ip_address','Total sales','Revenue']]
```

```python
AU_data['timestamp'].min(),AU_data['timestamp'].max()
```

```python
PRESENT = dt.datetime(2017,10,22)
AU_data['timestamp'] = pd.to_datetime(AU_data['timestamp'])
```

```python
PRESENT
```

```python
AU_data.head()
```

```python
rfm= AU_data.groupby('user ID').agg({'timestamp': lambda date: (PRESENT - date.max()).days,
                                        'ip_address': lambda num: len(num),
                                        'Revenue': lambda price: price.sum()})
```

```python
rfm.columns
```

```python
# Change the name of columns
rfm.columns=['recency','frequency','monetary']
```

```python
rfm
```

```python
rfm['recency'] = rfm['recency'].astype(int)
```

```python
rfm.head()
```

```python
rfm['r_quartile'] = pd.qcut(rfm['recency'], 4, ['1','2','3','4'])
rfm['f_quartile']=pd.qcut(rfm['frequency'].rank(method='first'),4 ,['4','3','2','1'])
rfm['m_quartile'] = pd.qcut(rfm['monetary'], 4, ['4','3','2','1'])
```

```python
rfm
```

```python
rfm.head()
```

```python
rfm['RFM_Score'] = rfm.r_quartile.astype(str)+ rfm.f_quartile.astype(str) + rfm.m_quartile.astype(str)
rfm.head()
```
