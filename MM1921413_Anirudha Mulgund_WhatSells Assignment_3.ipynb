```python
def below_ten(units_sold):
    if units_sold < 10:
        return 10
    else:
        return units_sold
    
```

```python
new_data2=new_data
```

```python
new_data2['units_sold'] = new_data2['units_sold'].apply(below_ten)
```

```python
new_data2['units_sold'].value_counts()
```

```python
sns.heatmap(new_data2.groupby(['units_sold'],as_index=False).mean().corr(), annot=True,cmap="YlGnBu")
plt.show()

#we observe that there is -0.9 negative correlation between units sold categories and price
```

```python
#Creating a newdataframe based on groups of units sold and considering their mean values across the data

type1=new_data2.groupby(['units_sold'],as_index=False).mean()
type1
```

```python
type2=type1[['units_sold','price']]
type2=type2.sort_values(by=['price'], ascending=False)
type2
```

```python
type2['cumm']=type2['price'].cumsum()
type2
```

```python
per=[]

for i in range(len(type2)):
    res=(type2['cumm'][i]/(125.923124))*100
    per.append(res)
    per.sort()

print(per)
```

```python
df2 = pd.DataFrame([per])
df2
```

```python
#compiled datset
df = pd.DataFrame(np.column_stack([type2,per]), columns= ['units_sold','price','cumm','cumm_per'])
df
```

```python
#Creating Pareto chart 
width = .35 # width of a bars
df[['price']].plot(kind='bar', width = width)
df['cumm_per'].plot(secondary_y=True)
ax = plt.gca()

plt.show()

#Interpretation:
# We observe that 80% of the products sold have unit price between Rs.8 to Rs.11 
```
